# Helper Functions

```python
topics = [
    "/joint_states.position[0]",
    "/joint_states.position[1]",
    "/joint_states.position[2]",
    "/static_joints_controller/commands.data[0]",
    "/static_joints_controller/commands.data[1]",
    "/static_joints_controller/commands.data[2]",
    "/joint_states.position[0]",
    "/joint_states.position[1]",
    "/joint_states.position[2]",
    "/joint_states.effort[0]",
    "/joint_states.effort[1]",
    "/joint_states.effort[2]",
]

def align_topics(df, topics=None, time_col="elapsed time", interpolate=False, fill_method="ffill"):
    """Pivot topics into columns and optionally fill/interpolate missing values."""
    if "timestamp" not in df.columns:
        raise ValueError("DataFrame must include a 'timestamp' column.")

    ts = pd.to_numeric(df["timestamp"], errors="coerce")
    if ts.isna().all():
        raise ValueError("All timestamps are NaN after conversion.")
    t0 = ts.min()

    work = df.copy()
    work["elapsed time"] = ts - t0

    if topics is not None:
        work = work[work["topic"].isin(topics)]

    if time_col not in work.columns:
        raise ValueError(f"Missing time column: {time_col}")

    wide = (
        work.pivot_table(
            index=time_col,
            columns="topic",
            values="value",
            aggfunc="first",
        )
        .sort_index()
    )

    if fill_method:
        wide = wide.fillna(method=fill_method)

    if interpolate:
        wide = wide.interpolate(method="linear", limit_direction="both")

    wide = wide.reset_index()

    # Ensure elapsed time is present even if aligning on timestamp
    if time_col != "elapsed time":
        wide["elapsed time"] = pd.to_numeric(wide[time_col], errors="coerce") - t0

    wide.columns.name = None
    return wide
```

```python
def export_data(dfs, out_dir, filenames=None):
    """Export a list of DataFrames to CSV files in out_dir."""
    from pathlib import Path

    out_dir = Path(out_dir)
    out_dir.mkdir(parents=True, exist_ok=True)

    if filenames is not None and len(filenames) != len(dfs):
        raise ValueError("filenames length must match dfs length.")

    for i, df in enumerate(dfs):
        name = filenames[i] if filenames is not None else f"df_{i}.csv"
        df.to_csv(out_dir / name, index=False)
```

```python
def effort2current(effort, motor_constant=0.091, effort_cols=None):
    """Convert motor effort (Nm) to current (A) using motor constant (Nm/A)."""
    import pandas as pd

    if motor_constant == 0:
        raise ValueError("motor_constant must be non-zero.")

    # DataFrame: scale effort columns in-place copy
    if isinstance(effort, pd.DataFrame):
        out = effort.copy()
        if effort_cols is None:
            effort_cols = [c for c in out.columns if c.startswith("/joint_states.effort[")]
        if not effort_cols:
            raise ValueError("No effort columns found to scale.")
        out[effort_cols] = out[effort_cols] / motor_constant
        return out

    # Series/array/scalar
    return effort / motor_constant
```
